```python
import numpy as np

linhas = int(input(("Digite o número de linhas da matriz: ")))
colunas = int(input("Digite o número de colunas da matriz: "))

#Define a variável img como um array (ou matriz) - Uso da biblioteca Numpy:
  #A matriz será preenchida inteiramente com o valor 1.
  #Tamanho da matriz definida pelo usuário

img = np.zeros((linhas,colunas))

#Apresenta o tamanho do array (img) produzido
print(f"Tamanho do array inserido: {img.shape}")

valor = int(input("Digite um valor para acrescentar a cada um dos elementos do array: "))
#Soma o valor inserido a cada um dos elementos do array inserido
cv2.add(img, valor)


```
