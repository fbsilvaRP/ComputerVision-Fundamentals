**Exemplos básicos utilizando a biblioteca Numpy**  

*   Definindo um array a partir da biblioteca Numpy
*   Adicionando um valor a cada um dos elementos do array



```python
import numpy as np

linhas = int(input(("Digite o número de linhas da matriz: ")))
colunas = int(input("Digite o número de colunas da matriz: "))

#Define a variável img como um array (ou matriz) - Uso da biblioteca Numpy:
  #A matriz será preenchida inteiramente com o valor 1.
  #Tamanho da matriz definida pelo usuário

img = np.zeros((linhas,colunas))

#OBS:Também é possível utilizar o método ones(apenas substituir zeros por ones), onde cada elemento será 1.

img2 = np.ones((linhas, colunas))

#Apresenta o tamanho do array (img) produzido
print(f"Tamanho do array inserido (array de zeros): {img.shape}")

#Apresenta o tamanho do array (img2) produzido
print(f"Tamanho do array inserido (array de 1s): {img2.shape}")

valor = int(input("Digite um valor para acrescentar a cada um dos elementos do array: "))

#Soma o valor inserido a cada um dos elementos do array inserido -- array de zeros
cv2.add(img, valor)

#Soma o valor inserido a cada um dos elementos do array inserido -- array de 1s
cv2.add(img2, valor)


```

**Download e extração dos arquivos necessários para o projeto**

```python
#Importação das bibliotecas necessárias...
import os
from zipfile import ZipFile
from urllib.request import urlretrieve

#Função para obter o arquivo da web
def baixar_e_extrair(url, diretorioParaSalvar):
  print(f"Baixando e extraindo arquivos...", end = "")

  #Baixa o arquivo compactado usando "urllib package"
  urlretrieve(url, diretorioParaSalvar)

  #Lidando com erros no funcionamento...
  try:
    #Extraindo o arquivo compactado usando o "zipfile package"
    with ZipFile(diretorioParaSalvar) as z:
      #Extraindo o arquivo compactado no mesmo diretório
      z.extractall(os.path.split(diretorioParaSalvar)[0])
    print("Feito.")
  except Exception as e:
    print("\nArquivo inválido.", e)

```
