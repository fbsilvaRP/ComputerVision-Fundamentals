```python
#Refazendo o código até...

import cv2
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
import random

url = "https://dcm.ffclrp.usp.br/images/mat.png"

resposta = urllib.request.urlopen(url)

imagem = np.asarray(bytearray(resposta.read()), dtype = "uint8")

imagem = cv2.imdecode(imagem, cv2.IMREAD_COLOR)

imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem)
plt.show()

#Aplicando o filtro...
novaImagem = imagem

linhas = novaImagem.shape[0]
colunas = novaImagem.shape[1]
tamanho = linhas * colunas

#Usei apenas 1 for...
for i in range(random.randint(0, tamanho)):
  #Definindo os pixels brancos
  y_branco = random.randint(0, linhas - 1)
  x_branco = random.randint(0, colunas - 1)
  novaImagem[y_branco][x_branco] = 255
  #Definindo os pixels pretos
  y_preto = random.randint(0, linhas - 1)
  x_preto = random.randint(0, colunas - 1)
  novaImagem[y_preto][x_preto] = 0

plt.imshow(novaImagem)
plt.show()
```
