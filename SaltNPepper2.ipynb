```python
#Última edição: 22 de abril de 2026, às 19:33
#Preciso:
  #Desenvolver uma nova lógica para reduzir o número de iterações para pintar os pixels...

#Importando a imagem...
import cv2
import urllib.request
import numpy
import IPython.display
import matplotlib.pyplot as plt
import random

#Endereço da imagem: deve ser inserido o endereço web da imagem
url = "https://dcm.ffclrp.usp.br/images/mat.png"

try:
  resposta = urllib.request.urlopen(url)
except:
  print("Erro ao obter o endereço inserido...Tente novamente.")

#-----------------------------------------------------------------------
#Transformação da imagem obtida:

imagem = numpy.asarray(bytearray(resposta.read()), dtype = "uint8")

imagem = cv2.imdecode(imagem, cv2.IMREAD_COLOR)
#-----------------------------------------------------------------------
#Exibindo a imagem...
print("Imagem sem a conversão de cores adequada:")
plt.imshow(imagem)
plt.show()
print("Imagem após a conversão:")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem)
plt.show()
#------------------------------------------------------------------------
#Aplicando o ruído "Sal e pimenta"

#Definindo as dimensões da imagem
linhas = imagem.shape[0]
colunas = imagem.shape[1]
canais = imagem.shape[2]

tamanho = linhas * colunas

print(f"A imagem obtida apresenta as dimensões: [{linhas}]X[{colunas}]")
print(f"A imagem apresenta {canais} canais de cores.")
print(f"A imagem apresenta {tamanho} pixels.")

#Definindo a quantidade de pixels da imagem a ser modificada:
PixelsBrancos = random.randint(0, tamanho)

novaImagem = imagem;

print(f"Foram escolhidos {PixelsBrancos} pixels para receberem o ruído.")

for i in range(PixelsBrancos):
  y = random.randint(0, linhas - 1)
  x = random.randint(0, colunas - 1)
  novaImagem[y][x] = 255

print(f"Nova imagem, apenas com sal. {PixelsBrancos} pixels foram modificados:")
plt.imshow(novaImagem)
plt.show()

```
